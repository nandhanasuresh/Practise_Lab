# **Salary** **Prediction**

```python
# Importing required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from word2number import w2n
```

```python
# Reading the data

df = pd.read_csv("/content/hiring.csv")
df
```

```python
# Cleaning the Data

df["experience"] = df["experience"].fillna("zero")
df
```

```python
# Cleaning the Data

df["experience"] = df["experience"].apply(w2n.word_to_num)
df
```

```python
# Cleaning the Data

df["test_score(out of 10)"] = df["test_score(out of 10)"].fillna(df["test_score(out of 10)"].median())
df
```

```python
df.info()
```

```python
# Converting test_score(out of 10) to integer type

df["test_score(out of 10)"] = df["test_score(out of 10)"].astype(int)
df
```

```python
df.describe()
```

```python
# Visualising the Data using Bar Chart

%matplotlib inline
plt.title('Graph')
plt.xlabel('experience')
plt.ylabel('salary($)')
plt.bar(df["experience"],df["salary($)"])
```

```python
# Visualising the Data using Scatter Diagram

%matplotlib inline
plt.title('Graph')
plt.xlabel('experience')
plt.ylabel('salary($)')
plt.scatter(df["experience"],df["salary($)"])
```

```python
# Visualising the Data using Normal Plot

%matplotlib inline
plt.title('Graph')
plt.xlabel('experience')
plt.ylabel('salary($)')
plt.plot(df["experience"],df["salary($)"])
```

```python
# Model initialisation

from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
# Fitting the data

input_data = df.drop(columns="salary($)")
output_data = df["salary($)"]
```

```python
input_data
```

```python
model.fit(input_data,output_data)
```

```python
# Predicting the Salary of an individual having 2 year experience,9 Test score and 6 Interview Score

model.predict([[2,9,6]])
```

```python
# Predicting the Salary of an individual having 12 year experience,10 Test score and 10 Interview Score

model.predict([[12,10,10]])
```

```python
from sklearn import metrics

# Predicting on the training data
predictions = model.predict(input_data)

# Calculating MAE (Mean Absolute Error)
mae = metrics.mean_absolute_error(output_data, predictions)
print("MAE:", mae)

# Calculating RMSE (Root Mean Squared Error)
rmse = np.sqrt(metrics.mean_squared_error(output_data, predictions))
print("RMSE:", rmse)

# Calculating Average Cost
average_cost = output_data.mean()
print("Average Cost:", average_cost)

# Calculating R-squared (R2)
r2 = metrics.r2_score(output_data, predictions)
print("R-Squared:", r2)
```
